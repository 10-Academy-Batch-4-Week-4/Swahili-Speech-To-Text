```python
def tran_loader(filename):
  name_to_text = {}
  with open (filename, encoding="utf-8")as f:
    f.readline()
    for line in f:
      name=line.split()[0]
      name=name.replace('\n','')
      name=name.replace(' ','')

      text=line.split()[1:]
      text = ' '.join(text)
      name_to_text[name]=text
    return name_to_text
```
